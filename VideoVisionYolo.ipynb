```python
from moviepy.editor import VideoFileClip

# Load the video clip
clip = VideoFileClip("VID_20170526_121819238.mp4")

# Get the original dimensions of the video
original_width = clip.size[0]
original_height = clip.size[1]

print(f"Original dimensions: {original_width} x {original_height}")
```

```python
original_width = 1920
original_height = 1080
target_width = 416
target_height = 416

# Calculate the scaling factors for width and height
width_scale = target_width / original_width
height_scale = target_height / original_height

# Determine the final scaling factor (choose the smaller of the two scales)
scale = min(width_scale, height_scale)

# Calculate the new dimensions
new_width = int(original_width * scale)
new_height = int(original_height * scale)

print(f"Original dimensions: {original_width} x {original_height}")
print(f"Resized dimensions: {new_width} x {new_height}")
```
